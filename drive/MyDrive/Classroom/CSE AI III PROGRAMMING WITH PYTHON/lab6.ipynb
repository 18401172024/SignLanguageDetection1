```python
input1=str(input())
input2=str(input())
result=""
max_length=max(len(input1), len(input2))
for i in range(max_length):
    if i < len(input1):
        result+=input1[i]
    if i < len(input2):
        result+=input2[i]
print(result)
```

```python
def expand_string(n):
    result=""
    i=0
    while i<len(n):
        letter=n[i]
        i += 1
        num_str=""
        while i<len(n) and n[i].isdigit():
            num_str += n[i]
            i += 1
        number = int(num_str)
        result += letter*number
    return result

n=input()
print(expand_string(n))
```
