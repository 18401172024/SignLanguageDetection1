```python
#swaping the value of two variable using assignment operator
x = int(input("the value of x is: "))
y = int(input("the value of y is: "))
x, y = y, x #swap the value
print("the result value of x is: ", x)
print("the result value of y is: ", y)
```

```python
#swaping the value of two variable using arithmetic operator
x = int(input("the value of x is: "))
y = int(input("the value of y is: "))
x = x + y 
y = x - y #swaping the value of y
x = x - y #swaping the value of x
print("the output value of x: ", x)
print("the output value of y: ", y)
```
