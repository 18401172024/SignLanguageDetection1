```python
#question no. 4
specified_limit=int(input("specified limit is "))
sum = 0
number = 1
sequence=""
#loop until the sum is greater than or equal to the limit
while sum+number <= specified_limit: #giving condition
    sum+=number
    if sequence == "":
        sequence = str(number)
    else:
        sequence += " + " + str(number)
    number += 1
print(f"the consecutive sum: {sequence} = {sum}")
```
