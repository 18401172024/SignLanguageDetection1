```python
n = int(input())
sum=0
for i in range(1, n+1):
    print("(",end='')
    for j in range(1, i+1):
        sum=sum+j
        if j!=n:
            if j!=i:
                print(j, end="+")
            else:
                print(j,end='')
        else:
            print(n,end=')=')
    if i!=n:
        print(")", end="+")
print(sum)
```

```python
def prime_factors(n):
    factors = []
    while n % 2 == 0:
        factors.append(2)
        n = n // 2
    for i in range(3, int(n**0.5) + 1, 2):
        while n % i == 0:
            factors.append(i)
            n = n // i
    if n > 2:
        factors.append(n)
    return factors
n = int(input())
if n > 1:
    print("Prime factors of", n, "are:", prime_factors(n))
else:
    print("Please enter a number greater than 1.")
```

```python
def factorial(n):
    if n==1 or n==0:
        return 1
    if n>1:
        return n*factorial(n-1)
result=factorial(5)
print(result)
```
